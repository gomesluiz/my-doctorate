```python
import pandas as pd
results = pd.read_csv('~/Workspace/issue-crawler/rscripts/results/20181104084742-long_live_bug-evaluation.csv')

# filling missing data
results['Acc_1'] = results.Acc_1.fillna(0)
results['Balanced_Acc'] = results.Balanced_Acc.fillna(results.Acc_0 / 2)

results.head(20)
```

```python
# slicing data
results = results[['Model', 'Threshold', 'Feature', 'Balanced_Acc']]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
   
def lineplot(data, feature):
    filtered_results = data[data['Feature'] == feature]
    filtered_results = filtered_results.pivot_table(index=['Threshold'], columns='Model', values='Balanced_Acc')
    filtered_results.plot(style=['r*-','bo-','y^-'], figsize=(15, 8), title='Balanced Accuracy Using %s Feature' % feature,
                     xticks=[4, 8, 16, 32, 64, 128, 256, 512], grid=True)

lineplot(results, 'Summary')
```

```python
lineplot(results, 'Description')
```
