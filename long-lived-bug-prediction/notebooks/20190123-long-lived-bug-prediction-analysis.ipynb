## 3. Clustering and Segmentation

```python
kclusters = 2
kmeans = KMeans(n_clusters = kclusters, random_state=0).fit(features.loc[:, 'action':'xml'])
features['cluster_label']  = kmeans.labels_
features['cluster_label'].value_counts().to_frame()
```

```python
features[['cluster_label', 'days_to_resolve']].groupby(['cluster_label']).agg(
    {
        'mean', 
        'median', 
        'std'
    }
)
```
